### b.Identificar outliers a partir de la distribució de cada atribut

**Un outlier es un valor atípic que s'allunya significativament de la resta de valors d'una variable**, per exepmle com es pot veure a la documentació si el depth dels diamants oscila entre (43 - 79) tindre un diamant amb depth = 15 es un valor molt allunyat de la mitja, que no significa que no sigui vàlid en segons quin context.

Els mètodes mes comuns son el **Z-score** i el **IQR**. El Z-score funciona noés en variables normalment distribuides en cambi l'IQR es millor quan no hi ha una simetria en les dades, per això farem servir IQR.

L'IQR és una mesura estadistica que ens diu quin és l'abast de la "zona central" de les dades. Utilitza per fer això els rangs interquartílics, si les dades estan distribuides entre el **25% - 75%** de les dades aleshores no son outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

columnes = ['carat', 'price', 'depth', 'x', 'y', 'z', 'table']

outliers_detectats = {}
estadistiques_outliers = {}

for col in columnes:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    límit_inferior = Q1 - 1.5 * IQR
    límit_superior = Q3 + 1.5 * IQR
    
    outliers = (df_train[col] < límit_inferior) | (df_train[col] > límit_superior)
    
    outliers_detectats[col] = df_train[outliers]
    
    estadistiques_outliers[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Límit inferior': límit_inferior,
        'Límit superior': límit_superior,
        'Outliers detectats': outliers.sum(),
        'Percentatge d\'outliers': (outliers.sum() / len(df_train)) * 100
    }

estadistiques_df = pd.DataFrame(estadistiques_outliers).T
print(estadistiques_df)

for col in columnes:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df_train[col])
    plt.title(f'Boxplot per a {col}')
    plt.show()
```

### d.Fer una taula amb els identificadors dels outliers, els seus atributs, i les raons per la qual cadascú es considera outlier

Per realitzar la taula només hem de repetir el procès anterior de detecció d'outliers en funció del preu que es el que volem trobar aleshores eliminar els que estiguin fora del rang IQR (aquest és el motiu)


```python
import pandas as pd

outliers_info = []

columnes = ['carat', 'price', 'depth', 'x', 'y', 'z', 'table']

# Iterar per cada columna i detectar outliers
for col in columnes:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    límit_inferior = Q1 - 1.5 * IQR
    límit_superior = Q3 + 1.5 * IQR
    
    # Identificar els outliers
    outliers = (df_train[col] < límit_inferior) | (df_train[col] > límit_superior)
    outliers_detectats = df_train[outliers]
    
    # Afegir la informació dels outliers a la llista
    for index, row in outliers_detectats.iterrows():
        raó = f"Valor de {col} fora del límit [ {límit_inferior}, {límit_superior} ]"
        outliers_info.append({
            "Identificador": row['id'],
            "Atribut": col,
            "Valor": row[col],
            "Raó": raó
        })

# Crear el DataFrame amb la informació dels outliers
taula_outliers = pd.DataFrame(outliers_info)
# Mostrar la taula dels outliers
print(taula_outliers)
```

### e.Decidir què fer amb els outliers i aplicar-ho als conjunts de dades.

Decidir què fer amb els outliers depèn de diversos factors com el context del projecte, tenim diverses opcions.

**Eliminar - Substituir - Mantenir - Transformar**

En el nostre cas al tindre no moltes dades es millor eliminar, ja que al substituir per la mitjana per exemple no tenim dades suficients com per aproximar una molt correcte, mantindre el outlier no tindria sentit i transformar-ho seria massa costós.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


def eliminar_outliers_iqr(df, columnas):
    """
    Elimina outliers de un DataFrame según el método IQR.
    Devuelve el DataFrame sin outliers y un diccionario con estadísticas.
    """
    estadistiques_outliers = {}
    mask_sense_outliers = pd.Series(True, index=df.index)

    for col in columnas:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        límit_inferior = Q1 - 1.5 * IQR
        límit_superior = Q3 + 1.5 * IQR

        estadistiques_outliers[col] = {
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'Límit inferior': límit_inferior,
            'Límit superior': límit_superior,
            'Outliers detectats': ((df[col] < límit_inferior) | (df[col] > límit_superior)).sum()
        }

        mask_sense_outliers &= (df[col] >= límit_inferior) & (df[col] <= límit_superior)

    df_sense_outliers = df[mask_sense_outliers].copy()
    return df_sense_outliers, estadistiques_outliers

columnes = ['carat', 'price', 'depth', 'x', 'y', 'z', 'table']

df_train, stats_train = eliminar_outliers_iqr(df_train, columnes)
df_test, stats_test = eliminar_outliers_iqr(df_test, columnes)

print(f"✅ Tamaño train sin outliers: {df_train.shape[0]} filas")
print(f"✅ Tamaño test sin outliers: {df_test.shape[0]} filas")

# Mostrar estadísticas (opcional)
print("📊 Estadísticas train:")
print(pd.DataFrame(stats_train).T)

print("\n📊 Estadísticas test:")
print(pd.DataFrame(stats_test).T)

```

### f. Convertir els atributs categòrics en numèrics.

Aquest pas es necesari per aplicar models a les nostres dades i tenint en compte l'ordre de millor a pitjor apliquem aquestes transformacions en atributs numèrics.

```python
cut_mapping = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}
color_mapping = {
    'J': 1,
    'I': 2,
    'H': 3,
    'G': 4,
    'F': 5,
    'E': 6,
    'D': 7
}
clarity_mapping = {
    'I1': 1,
    'SI2': 2,
    'SI1': 3,
    'VS2': 4,
    'VS1': 5,
    'VVS2': 6,
    'VVS1': 7,
    'IF': 8
}

df_train['cut'] = df_train['cut'].map(cut_mapping)
df_train['color'] = df_train['color'].map(color_mapping)
df_train['clarity'] = df_train['clarity'].map(clarity_mapping)

df_test['cut'] = df_test['cut'].map(cut_mapping)
df_test['color'] = df_test['color'].map(color_mapping)
df_test['clarity'] = df_test['clarity'].map(clarity_mapping)

print(df_train.head())

```

### g. Estandaritzar tots els atributs per separat, excepte x, y, z i price

Estandaritzar els atributs consisteix en **transformar els valors de cada atribut per a que la seva mitjana sigui 0 i la desviació estàndard 1.** Això es fa amb aquesta formula:

**z = x (valor del atribut) - 𝜇 (mitjana del atribut) / σ (desviació estàndard)**

Això es fa perque en Machine Learning molts algorismes funcionen millor si les variables tenen escales comparables, sino podríem tindre valors grans que dominen sobre altres.

En aquest cas no s'estandaritza les mesures físiques ni el preu, el preu és el target aleshores no volem estandaritzar el que es vol predir

```python
from sklearn.preprocessing import StandardScaler

atributs_a_estandaritzar = ['carat', 'depth', 'table', 'cut', 'color', 'clarity']

print("Mitjanes abans d'estandaritzar:")
print(df_train[atributs_a_estandaritzar].mean())

print("\nDesviacions abans després d'estandaritzar:")
print(df_train[atributs_a_estandaritzar].std())

scaler = StandardScaler()
df_train[atributs_a_estandaritzar] = scaler.fit_transform(df_train[atributs_a_estandaritzar])
df_test[atributs_a_estandaritzar] = scaler.transform(df_test[atributs_a_estandaritzar])

print("Mitjanes després d'estandaritzar:")
print(df_train[atributs_a_estandaritzar].mean())

print("\nDesviacions estàndard després d'estandaritzar:")
print(df_train[atributs_a_estandaritzar].std())

print(df_train.head())
```

### h.Escala x, y i z entre -1 i 1, però de forma conjunta, no cada atribut per separat.

Per escalar els atributs de forma cojunta entre -1 i 1 hem utilitzat una tècnica d'escalat basada en el mínim i màxim conjunt per a les tres variables

*scaled_value = (2*(value - min_value(x,y,z)) / (max_value(x,y,z) - min_value(x,y,z)))-1*

```python
xyz_all = pd.concat([
    df_train[['x', 'y', 'z']],
    df_test[['x', 'y', 'z']]
], ignore_index=True)

# Convertim a array pla i trobem min i max globals
xyz_values = xyz_all.values.flatten()
min_xyz = xyz_values.min()
max_xyz = xyz_values.max()

print(f"Escalat global: min = {min_xyz}, max = {max_xyz}")

# Funció d’escalat conjunt
def escala_xyz(df, min_val, max_val):
    return 2 * (df[['x', 'y', 'z']] - min_val) / (max_val - min_val) - 1

# Apliquem al train i al test
df_train[['x', 'y', 'z']] = escala_xyz(df_train, min_xyz, max_xyz)
df_test[['x', 'y', 'z']] = escala_xyz(df_test, min_xyz, max_xyz)

display(df_train.head())
display(df_train.head())
```

### i. Calcular el logaritme dels preus i guardar-lo en un nou atribut log_price, i aplicar un escalament lineal entre 0.1 i 0.9

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

def scale_minmax(series, min_val, max_val, a=0.0, b=1.0):
    return a + ((series - min_val) / (max_val - min_val)) * (b - a)

# Calcular el logaritme del preu i afegir-ho com a nova columna
df_train['log_price'] = np.log(df_train['price'])

min_log_price = df_train['log_price'].min()
max_log_price = df_train['log_price'].max()

df_train['log_price'] = scale_minmax(df_train['log_price'], min_log_price, max_log_price, a=0.1, b=0.9)

df_test['log_price'] = np.log(df_train['price'])
df_test['log_price'] = scale_minmax(df_test['log_price'], min_log_price, max_log_price, a=0.1, b=0.9)

# Comprovar els resultats
print(df_train[['log_price', 'price']])
```

### j.Construir un nou atribut high_price que valgui 0 si el preu està estrictament per sota dels $2,500 i 1 en cas contrari.

```python
df_train['high_price'] = (df_train['price'] >= 2500).astype(int)
df_test['high_price'] = (df_test['price'] >= 2500).astype(int)

# Comprovar els resultats
print(df_train[['price', 'high_price']])
print(df_test[['price', 'high_price']])
```

# 3.Predicció del preu

Ara hem de predir el preu del diamant a partir de les seves caraterístiques. Realitzarem totes les tasques anomenades per cada model.

### i. Regresió Multilineal (MLR)

La regresió multilineal és un model que busca ajustar una línia en moltes dimensions per predir un valor continu a partir de variables independents.

Primer haurem de fer Cross-Validation quue  ens ajudarà a evaluar el redniment real d'un model fent particions de les dades per evitar que el model depengui massa de com s'ha dividit el conjunt.

També calcularem el MAPE (Mean Absolute Percentage Error) que medeix l'error percentual mitjà entre la predicció i el valor real.

Abans declararem les funcions i importarem els paquets necesaris per realitzar les operacions de transformació de logs -> price i MAPE

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor
import numpy as np

# Funcio per pasar de log_escalat a log_price
def inverse_scale_log_price(scaled_log_p):
    return min_log_price + (scaled_log_p - 0.1) * (max_log_price - min_log_price) / 0.8

#Funció per pasar de log_price a price
def log_to_price(log_p):
    return np.exp(log_p)

# Fórmula MAPE
def mape(y_true, y_pred):
    return 100 * np.mean(np.abs((y_pred - y_true) / y_true))

# Agafem la columna de price per comparar les prediccions
price_real = df_test['price'].values
```

Per entrenar als models necesitem treure la columna id que no aporta cap dada, el price, log_price i high_price ja que son les respostes a la nostra predicció

```python
x_train = df_train.drop(columns=['id', 'price', 'log_price', 'high_price'])
y_train = df_train['log_price']

x_test = df_test.drop(columns=['id', 'price', 'log_price', 'high_price'])
y_test = df_train['log_price']
```

Declarem el model i realitzem el Cross-Validation

```python
mlr_model = LinearRegression()

cv_scores_mlr = cross_val_score(mlr_model, x_train , y_train, cv=5, scoring='neg_mean_absolute_percentage_error')
print(f'MAPE promig en la validació creuada: {-cv_scores_mlr.mean()*100} %')
```

En el cas del MLR no hi ha paràmetres a ajustar com si ocurrirà amb altres models. Per osi que hem d'entrenar al model definitiu amb el conjunt d'entrenament per després utilitzar-ho amb el de test

```python
#Entrenem al model amb les dades de diamondts-train.csv

mlr_model.fit(x_train, y_train)

# Revertim el log escalat a price i calculem MAPE
mlr_log_price_pred = mlr_model.predict(x_test)
mlr_price_predict = log_to_price(inverse_scale_log_price(mlr_log_price_pred))
mape_mlr = mape(price_real, mlr_price_predict)

print(mape_mlr);
```

### ii. K-Nearest Neighbors (k-NN)

k-NN prediu el valor del preu a parir dels seus veins més propers basat en característiques a través de distancies. Es a dir trobant els veïns més propers

Primer fem un Grid Search fent cross validation per trobar el valor de K més adecuat per les nostres dades.

```python
knn = KNeighborsRegressor()
param_grid = {'n_neighbors': list(range(3, 21))}

grid_knn = GridSearchCV(knn, param_grid, cv=5, scoring='neg_mean_absolute_percentage_error')
grid_knn.fit(x_train, y_train)

best_k = grid_knn.best_params_['n_neighbors']

print(f"Millor k: {best_k}, MAE mitjà (CV): {-grid_knn.best_score_:.5f}")
```

Ara tenint el compte el millor K per les dades ja podem fer les prediccions del model i calcular el MAPE

```python
knn_optim = KNeighborsRegressor(best_k)
knn_optim.fit(x_train, y_train)

knn_log_price_predict = knn_optim.predict(x_test)

knn_price_predict = log_to_price(inverse_scale_log_price(knn_log_price_predict))

mape_knn = mape(price_real, knn_price_predict)
print(mape_knn)
```

### iii. Multilayer Neural Network amb Back-Propagation

Una Multilayer Neural Network està formada per una capa d'entrada, capes ocultes i una de sortida, cada neuraona aplica una transformació lineal seguida d'una funció d'activació. L'entrenarem utilitzant backpropagation, un algoritme per ajustar els pesos i minimitzar l'error.

```python
mlp = MLPRegressor(random_state=42, max_iter=1000)

param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (100, 50), (50, 25)],
    'activation': ['relu', 'tanh'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate_init': [0.001, 0.01]
}

grid_mlp = GridSearchCV(mlp, param_grid, cv=5, scoring='neg_mean_absolute_percentage_error', n_jobs=-1, verbose=2)
grid_mlp.fit(x_train, y_train)

print("Mejores hiperparámetros:", grid_mlp.best_params_)
print(f"MAPE medio (CV): {-grid_mlp.best_score_:.4f}")
```

```python
mlp_log_price_pred = grid_mlp.predict(x_test)

mlp_price_predict = log_to_price(inverse_scale_log_price(mlp_log_price_pred))

mape_mpl = mape(price_real, mlp_price_predict)
print(mape_mpl)
```

### Anàlisi de resultats

```python
plt.figure(figsize=(6, 5))
model_names = ['MLR', 'k-NN', 'MPL']
mape_values = [mape_mlr, mape_knn, mape_mpl]

plt.bar(model_names, mape_values, color=['skyblue', 'orange', 'green'])
plt.title("Comparació del MAPE entre models")
plt.ylabel("MAPE (%)")
plt.ylim(0, max(mape_values) * 1.2)

# Mostrar los valores encima de las barras
for i, value in enumerate(mape_values):
    plt.text(i, value + 0.2, f"{value:.2f}%", ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)

plt.scatter(price_real, mlr_price_predict, alpha=0.5, color='skyblue')
plt.plot([price_real.min(), price_real.max()], [price_real.min(), price_real.max()], 'r--')
plt.title("Regressió Lineal")
plt.xlabel("Preu real")
plt.ylabel("Preu predit")

plt.subplot(1, 3, 2)
plt.scatter(price_real, knn_price_predict, alpha=0.5, color='orange')
plt.plot([price_real.min(), price_real.max()], [price_real.min(), price_real.max()], 'r--')
plt.title("K-Nearest Neighbors")
plt.xlabel("Preu real")
plt.ylabel("Preu predit")

plt.subplot(1, 3, 3)
plt.scatter(price_real, mlp_price_predict, alpha=0.5, color='green')
plt.plot([price_real.min(), price_real.max()], [price_real.min(), price_real.max()], 'r--')
plt.title("MLP Neural Network")
plt.xlabel("Preu real")
plt.ylabel("Preu predit")

plt.tight_layout()
plt.show()
```

Segons els resultats podem veure el K-Nearest neighbors tendeix a donar un preu dels diamants més alt, els MLP més baix i el MLR es queda entremig dels dos models.
